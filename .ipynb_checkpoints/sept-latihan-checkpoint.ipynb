# DATA CUSTOMERS

```python
customers = pd.read_csv('./data/customers.csv')
customers.head()
```

```python
customers.info()
```

```python
customers.describe()
```

```python
customers.duplicated().sum()
```

```python
customers.drop_duplicates(inplace=True)
```

```python
# customers[customers.age == customers.age.max()]
# customers.loc[customers.age == customers.age.max()]
customers['age'].nlargest(3)
```

```python
customers.loc[[215, 967], 'age'] = [50, 70]
customers.loc[[215, 967]]
```

```python
customers.isna().sum()
```

```python
# customers.gender.isna().sum()
customers.gender.value_counts()
```

```python
customers.loc[customers.gender.isna(), 'gender'] = 'Prefer not to say'
```

```python
customers.info()
```

```python
customers = customers.astype({
    'customer_name': 'string',
    'gender': 'category',
    'home_address': 'string',
    'city': 'string',
    'state': 'string',
    'country': 'string',
})
```

```python
customers[['code_address', 'street_address']] = customers.home_address.str.title().str.strip().str.split(n=1, expand=True)
customers['code_address'] = customers['code_address'].astype('uint64')
```

```python
products = pd.read_csv('./data/products.csv')
products.head()
```

```python
products.info()
```

```python
products.duplicated().sum()
```

```python
products.drop_duplicates(inplace=True)
```

```python
products.describe(include='all')
```

```python
# products['product_type'].unique()
# products['product_name'].unique()
# products['colour'].unique()
products['size'].unique()
```

```python
products['size'] = products['size'].str.strip().str.upper()
products['product_name'] = products['product_name'].str.strip().str.title()
products['product_type'] = products['product_type'].str.strip().str.title()
```

```python
products = products.astype({
     'product_id' : 'uint64',
     'product_type' : 'category',
     'product_name' : 'string',
     'size' : 'category',
     'colour' : 'category',
     'price' : 'uint64',
     'quantity' : 'uint32',
     'description' : 'string'
})
```

```python
products.describe(include='all')
```
